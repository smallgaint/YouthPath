# 02 정책 검색 품질 확인

인덱싱 후 테스트 쿼리별 상위 5개 결과를 수동 검수한다.

```python
from core.rag import search

queries = [
    '서울 청년 주거 지원',
    '30살 이하 창업 지원금',
    '대학생 학자금 대출',
    '취업 준비 청년 지원',
    '청년 복지 상담 지원',
]

for query in queries:
    print('\n##', query)
    for item in search('policies', query, k=5):
        print(item['metadata'].get('title'), item['score'])
```
